```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('AAPL.csv',parse_dates=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data['Date'] = pd.to_datetime(data['Date'],format="%Y-%m-%d")
```

```python
data.isna().sum()
```

```python
def createLagData(inpDf, numShift, colName):
    ldf = inpDf.copy()
    for i in range(numShift):
        ldf[colName+str(i+1)] = ldf[colName].shift(i+1)
    return(ldf)
```

```python
data = data[['Date','Close']]
```

```python
data.head()
```

```python
data = createLagData(data,60,'Close')
```

```python
data.head()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data.tail()
```

```python
df_train = data.head(data.shape[0]-30)
df_test = data.tail(30)
```

```python
X_train = df_train.drop(['Date','Close'],axis=1)
X_test =  df_test.drop(['Date','Close'],axis=1)
```

```python
y_train = df_train['Close']
y_test = df_test['Close']
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
X_train_sc = np.reshape(X_train_sc, (X_train_sc.shape[0], X_train_sc.shape[1],1))
```

```python
X_test_sc = np.reshape(X_test_sc, (X_test_sc.shape[0], X_test_sc.shape[1],1))
```

```python
X_train_sc.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
```

```python
rnn_model = Sequential()
```

```python
rnn_model.add(LSTM(units=128, 
                   return_sequences=True, 
                   input_shape=(X_train_sc.shape[1],1)))
rnn_model.add(Dropout(0.2))
```

```python
rnn_model.add(LSTM(units=64, return_sequences=True))
rnn_model.add(Dropout(0.2))
```

```python
rnn_model.add(LSTM(units=32, return_sequences=True))
rnn_model.add(Dropout(0.2))
```

```python
rnn_model.add(LSTM(units=16, return_sequences=False))
rnn_model.add(Dropout(0.2))
```

```python
rnn_model.add(Dense(units = 1))
```

```python
rnn_model.compile(optimizer='adam',
                  loss = 'mse',
                  metrics = [tf.keras.metrics.MeanAbsolutePercentageError()]
                 )
```

```python
rnn_model.summary()
```

```python
rnn_model.fit(X_train_sc,y_train, validation_data = (X_test_sc, y_test),epochs=500, batch_size = 32)
```

```python
y_pred = rnn_model.predict(X_test_sc)
```
